# Completando la cantidad de municipios tabla por mes 2005 -2021

```python
import pandas as pd
```

```python
# Carga el CSV en un DataFrame
tabla_mes = pd.read_csv('completando/ejemplo/2005_sospecha-diagnostica_difteria_mes.csv')
tabla_mes
```

```python
# Elimina las filas duplicadas en el DataFrame original
tabla_mes = tabla_mes.drop_duplicates(subset=['codigo_municipio', 'mes'])
tabla_mes
```

```python
# Crea un DataFrame con códigos de municipio únicos y sus nombres correspondientes
df_municipios_unicos = tabla_mes[['codigo_municipio', 'municipio']].drop_duplicates()
df_municipios_unicos
```

```python
#cargando el CSV con la lista de departamentos y sus municipios

todos_municipios = pd.read_csv('resources/municipios_snis_corregido.csv')##tabla editada luego de eliminar muncipios repetidos
todos_municipios
```

```python
# Renombrar las columnas para que coincidan con la tabla incompleta
todos_municipios = todos_municipios.rename(columns={'COD_MUNICIPIO': 'codigo_municipio', 'MUNICIPIO': 'municipio'})
#todos_municipios = todos_municipios.rename(columns={'DEPARTAMENTO': 'DEPARTAMENT', 'COD_MUNICIPIO': 'code_municipality', 'MUNICIPIO': 'municipality'})
todos_municipios
```

```python
# Reorganizar las columnas para que "departamento" esté en la primera posición
todos_municipios = todos_municipios[['DEPARTAMENTO', 'codigo_municipio', 'municipio']]
#todos_municipios = todos_municipios[['DEPARTAMENT', 'code_municipality', 'municipality']]
todos_municipios[:3]
```

```python
# Crea un DataFrame con todas las fechas únicas
df_fechas_completas = tabla_mes[['mes']].drop_duplicates()
df_fechas_completas
```

```python
# Realiza una combinación (merge) entre los DataFrames de municipios y fechas
#usando 'cross' se produce una multiplicacion cartesiana donde se juntan ambos DF  sin necesidad de columnas en comùn
df_combinado = pd.merge(todos_municipios, df_fechas_completas, how='cross')
df_combinado
```

```python
# Realiza una fusión (merge) con el DataFrame original para obtener los valores correspondientes
df_completo = pd.merge(df_combinado, tabla_mes, on=['DEPARTAMENTO','codigo_municipio','municipio','mes'], how='left')
df_completo
```

```python

df_completo['grupo'] = tabla_mes['grupo'].unique()[0]  # Asigna el valor único de 'grupo' en tabla_incompleta
df_completo['variable'] = tabla_mes['variable'].unique()[0]  # Asigna el valor único de 'variable' en tabla_incompleta
# Rellena las columnas seleccionadas a partir del la sexta con 0 y las convierte a tipo entero
df_completo.iloc[:, 6:] = df_completo.iloc[:, 6:].fillna(0).astype(int)
```

```python
df_completo
```

```python
# Guardar el DataFrame resultante como un archivo CSV
df_completo.to_csv('completando/tabla_mes_difteria_2005.csv', index=False)
```
