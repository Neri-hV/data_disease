```python
# Carga el CSV en un DataFrame
tabla_incompleta = pd.read_csv('completando/ejemplo/2001_sospecha-diagnostica_tosferina.csv')
tabla_incompleta[:5]
```

```python
todos_municipios = pd.read_csv('resources/municipios_snis_corregido.csv')##tabla editada luego de eliminar muncipios repetidos
todos_municipios[:5]
```

```python
todos_municipios = todos_municipios.rename(columns={'DEPARTAMENTO': 'DEPARTAMENT', 'COD_MUNICIPIO': 'code_municipality', 'MUNICIPIO': 'municipality'})
todos_municipios = todos_municipios[['DEPARTAMENT', 'code_municipality', 'municipality']]
todos_municipios
```

```python
tabla_incompleta = tabla_incompleta.rename(columns={'codigo_municipio':'code_municipality', 'municipio' :'municipality', 'grupo':'group' })
tabla_incompleta[:3]
```

```python
# Convertir la columna "mes" a formato de fecha
tabla_incompleta['mes'] = pd.to_datetime(tabla_incompleta['mes'])
# Agregar columnas "año" y "departamento" basadas en la columna "mes"
tabla_incompleta['year'] = tabla_incompleta['mes'].dt.year
# Sumar los valores mensuales por municipio y año
tabla_sumada = tabla_incompleta.groupby(['code_municipality', 'municipality', 'group', 'variable','year']).sum().reset_index()
```

```python
tabla_sumada
```

## traduciendo grupo y variable

```python
valores_unicos = tabla_sumada["group"].unique()
valores_unicos
```

```python
# Creamos un diccionario que mapea los valores originales a los valores traducidos
diccionario = {
    "sospecha diagnostica": "diagnostic suspicion"
}
# Aplicamos la función de traducción a la columna
tabla_sumada["group"] = tabla_sumada["group"].apply(lambda x: diccionario.get(x, x))
```

```python
valores_unicos = tabla_sumada["variable"].unique()
valores_unicos
```

```python
# Creamos un diccionario que mapea los valores originales a los valores traducidos
diccionario = {
    "tosferina": "whooping cough"
}
# Aplicamos la función de traducción a la columna
tabla_sumada["variable"] = tabla_sumada["variable"].apply(lambda x: diccionario.get(x, x))
```

# agregando muncipios faltantes

```python
tabla_completa = todos_municipios.merge(tabla_sumada, on=['code_municipality', 'municipality'], how='left')
tabla_completa['group'] = tabla_sumada['group'].unique()[0]  # Asigna el valor único de 'grupo' en tabla_incompleta
tabla_completa['variable'] = tabla_sumada['variable'].unique()[0]  # Asigna el valor único de 'variable' en tabla_incompleta
tabla_completa['year'] = tabla_sumada['year'].unique()[0]  # Asigna el valor único de 'año' en tabla_incompleta
tabla_completa.iloc[:, 6:] = tabla_completa.iloc[:, 6:].fillna(0).astype(int)

tabla_completa
```
