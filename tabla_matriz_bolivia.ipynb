```python
import pandas as pd
import os
```

```python
# Directorio que contiene los archivos CSV
directorio = 'C:/Users/USER/Documents/Boliva_entrega/Supporting_information_5/tabla_matriz'


# Crear una lista para almacenar las quintas columnas
quintas_columnas = []

# Iterar a través de los archivos CSV en el directorio
for filename in os.listdir(directorio):
    if filename.endswith('.csv'):
        filepath = os.path.join(directorio, filename)

        # Leer el archivo CSV
        df = pd.read_csv(filepath)
        # Ordenar el DataFrame por 'DEPARTAMENT' y 'municipality'
        df.sort_values(by=['DEPARTAMENT', 'municipality'], inplace=True)

        # Obtener la quinta columna (columna de enfermedad)
        quinta_columna = df.iloc[:, 4]

        # Agregar la quinta columna a la lista
        quintas_columnas.append(quinta_columna)

# Concatenar las quintas columnas en un solo DataFrame
quinta_columna_final = pd.concat(quintas_columnas, axis=1)

# Obtener las columnas comunes del primer archivo CSV
primer_csv = pd.read_csv(os.path.join(directorio, os.listdir(directorio)[0]))
columnas_comunes = primer_csv[['DEPARTAMENT', 'code_municipality', 'municipality', 'year']]

# Concatenar la quinta columna con las columnas comunes
tabla_final = pd.concat([columnas_comunes, quinta_columna_final], axis=1)

# Guardar el DataFrame final en un archivo CSV
tabla_final.to_csv('tabla_final_completa.csv', index=False, float_format='%.0f')
```
